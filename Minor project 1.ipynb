### RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
param_dist={'n_estimators':[int(x) for x in np.linspace(start=100,stop=600,num=6)],
            'max_features':['auto'],
            'min_samples_leaf':[1,2,3,4],
            'max_depth':[int(x) for x in np.linspace(2,30,10)],
            'min_samples_split':[int(x) for x in np.linspace(2,15,5)]}
```

```python
def hyper(rfr,param,X,y):
    random=RandomizedSearchCV(estimator=regr,param_distributions=param_dist,n_jobs=-1,cv=5,n_iter=20)
    random.fit(X,y)
    ht_params=random.best_params_
    ht_score=random.best_score_
    return ht_params,ht_score
```

```python
parameters,scores=hyper(regr,param_dist,X,y)
```

```python
regr = RandomForestRegressor(n_estimators= 100,max_depth= 26, min_samples_split=5,min_samples_leaf=1,max_features='auto')
regr.fit(X_train, y_train)
```

```python
regr.score(X_train,y_train)
```
